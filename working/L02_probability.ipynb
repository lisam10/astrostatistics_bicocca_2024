```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import array as arr
import random as r

wins_s = 0
wins_c = 0
wins_l = 0

runs_n = 10000
#switcher cicle
for i in range(runs_n):
    #generate 3 doors and 1 car
    doors = np.array ([0, 0, 0])
    car_n = r.randint(0, 2)
    doors[car_n] = 1
    
    #switcher
    switcher = np.array ([0, 0, 0])
    swit_n = r.randint(0, 2)
    switcher[swit_n] = 1

    #door opening
    open_n = r.randint(0, 2)
    while (open_n == car_n or open_n == swit_n):
        open_n = r.randint(0, 2)

    doors[open_n] = 2
    switcher[open_n] = 2

    for x in switcher:
        if (switcher[x] == 0):
            swit_n = x

    if (swit_n == car_n): wins_s = wins_s + 1     

#conservative cicle
for i in range(runs_n):
    #generate 3 doors and 1 car
    doors = np.array ([0, 0, 0])
    car_n = r.randint(0, 2)
    doors[car_n] = 1
    
    #conservative
    cons = np.array ([0, 0, 0])
    cons_n = r.randint(0, 2)
    cons[cons_n] = 1

    #door opening
    open_n = r.randint(0, 2)
    while (open_n == car_n or open_n == cons_n):
        open_n = r.randint(0, 2)

    doors[open_n] = 2

    if (cons_n == car_n): wins_c = wins_c + 1

#lucky guy
for i in range(runs_n):
    #generate 3 doors and 1 car
    doors = np.array ([0, 0, 0])
    car_n = r.randint(0, 2)
    doors[car_n] = 1
    
    #door opening
    open_n = r.randint(0, 2)
    while (open_n == car_n):
        open_n = r.randint(0, 2)

    doors[open_n] = 2
    switcher[open_n] = 2
    
    #third guy
    third_n = r.randint(0, 2)
    while (third_n == open_n): third_n = r.randint(0, 2)

    if (third_n == car_n): wins_l = wins_l + 1     

print("\n", "{:.2f}".format(wins_s/runs_n))
print("\n", "{:.2f}".format(wins_c/runs_n))
print("\n", "{:.2f}".format(wins_l/runs_n))

#now we repeat the process with 100 doors and opening 98 of them
#switcher cicle


#generate 100 doors and 1 car



```
