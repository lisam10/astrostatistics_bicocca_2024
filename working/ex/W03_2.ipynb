### 2. Kicking horses

A famous early  application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by [Bortkiewicz](https://www.wikiwand.com/en/Ladislaus_Bortkiewicz) in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

- Plot the resulting probability distribution. Careful with the normalization
- How does it look like? Again try to fit it by hand.

(Hint: think about sample mean and sample variance)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
groups = np.array([109, 65, 22, 3, 1])
deaths = np.array([0, 1, 2, 3, 4])

#-- Normalizing by using the total # of groups
groups = groups / np.sum(groups)
```

```python
plt.plot(deaths, groups, 'o-')
plt.xlabel('Number of deaths')
plt.ylabel('% of groups')
plt.show()
```

We will use the Poisson distribution, as we need the probability of k events given the mean $\mu$. If we have a sample of the entire population, we can use the sample mean to estimate $\mu$.

```python
sample_mean = np.mean(deaths)
weighted_mean = np.average(deaths, weights = groups)
pdf = scipy.stats.poisson(sample_mean)
pdf_w = scipy.stats.poisson(weighted_mean)

plt.plot(deaths, pdf.pmf(deaths), label = r'$\mu$ = ' + str(sample_mean))
plt.plot(deaths, pdf_w.pmf(deaths), label = r'$\mu_\mathrm{weighted}$ = ' + f'{weighted_mean:.2f}')
plt.plot(deaths, groups, 'o')
plt.xlabel('Number of deaths')
plt.ylabel('% of groups')
plt.title('Fit with Poisson distributions')
plt.legend()
plt.show()
```
