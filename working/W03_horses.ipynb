# Kicking horses
A famous early  application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by [Bortkiewicz](https://www.wikiwand.com/en/Ladislaus_Bortkiewicz) in 1898. He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:


 Number of deaths: 0   1   2   3   4
 
 Number of groups: 109 65 22 3 1
 
 
- Plot the resulting probability distribution. Careful with the normalization
- How does it look like? Again try to fit it by hand

(Hint: think about sample mean and sample variance)

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import poisson

deathsn = np.array([0, 1, 2, 3, 4])
groupsn = np.array([109.0, 65.0, 22.0, 3.0, 1.0])

tot_groups = 200

for x in range(len(groupsn)):
    groupsn[x] = groupsn[x]/tot_groups #dividing the y data by the sum to normalise it to 1
    
plt.plot(deathsn, groupsn, 'o-')
plt.xlabel("number of deaths")
plt.ylabel("number of corps")


mu = int(deathsn.mean())
plt.plot(deathsn, poisson.pmf(deathsn, mu))
```
